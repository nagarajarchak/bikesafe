```python
import geopandas as gpd
import pandas as pd

# Path to the GeoJSON file
geojson_path = 'geojsonsub.geojson'

# Read the GeoJSON file into a GeoDataFrame
gsubdf = gpd.read_file(geojson_path)
```

```python
# Path to the GeoJSON file
geojson_path_1 = 'crash_vic.geojson'

# Read the GeoJSON file into a GeoDataFrame
gvicdf = gpd.read_file(geojson_path_1)

```

```python
gvicdf.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
gvicdf.head()
```

```python
temp = gvicdf[gvicdf['LGA_NAME']=="MELBOURNE"]
```

```python
temp.head()
```

```python
temp.columns
```

```python
temp = temp.loc[:,['ACCIDENT_NO','ACCIDENT_DATE','ACCIDENT_TYPE','SEVERITY','LGA_NAME','geometry']]
```

```python
temp.tail()
```

```python
# Convert the 'ACCIDENT_DATE' column to datetime format
temp['ACCIDENT_DATE'] = pd.to_datetime(temp['ACCIDENT_DATE'], format='%Y%m%d')
```

```python
temp.head()
```

```python
gsubdf.head()
```

```python
temp2= gsubdf.loc[:,['vic_loca_2','geometry']]
```

```python
temp2=temp2.rename(columns={'vic_loca_2': 'suburb'})
```

```python
temp2.head()
```

```python
# Perform spatial join to find which suburb each point lies within
joined_gdf = gpd.sjoin(temp, temp2, how="left", predicate='within')
```
