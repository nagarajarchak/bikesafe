```python
crash_data=pd.read_csv("crash.csv")
```

```python
import pandas as pd
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import plotly.graph_objs as go

# Aggregate data for choropleth map
suburb_counts = crash_data.groupby('suburb').size().reset_index(name='ACCIDENT_COUNT')

# Combine with lat/lon data
suburb_counts = suburb_counts.merge(crash_data[['suburb', 'LATITUDE', 'LONGITUDE']].drop_duplicates(), on='suburb')

# Create the Dash app
app = Dash(__name__)

# Layout of the app
app.layout = html.Div([
    html.H1("Accidents by Suburb", style={'text-align': 'center'}),
    dcc.Graph(id='choropleth-map'),
    dcc.Graph(id='bar-chart')
])

# Callback for updating the choropleth map
@app.callback(
    Output('choropleth-map', 'figure'),
    Input('choropleth-map', 'clickData')
)
def update_map(clickData):
    fig = px.scatter_mapbox(
        suburb_counts,
        lat="LATITUDE",
        lon="LONGITUDE",
        size="ACCIDENT_COUNT",
        color="ACCIDENT_COUNT",
        hover_name="suburb",
        hover_data={"LATITUDE": False, "LONGITUDE": False},
        color_continuous_scale="Viridis",
        zoom=8,
        mapbox_style="carto-positron"
    )
    fig.update_layout(
        title="Accidents by Suburb",
        mapbox=dict(center={"lat": -37.8136, "lon": 144.9631}, zoom=10),
        margin={"r":0,"t":0,"l":0,"b":0},
        height=600
    )
    return fig

# Callback for updating the bar chart
@app.callback(
    Output('bar-chart', 'figure'),
    Input('choropleth-map', 'clickData')
)
def update_bar_chart(clickData):
    if clickData:
        suburb = clickData['points'][0]['hovertext']
        filtered_df = crash_data[crash_data['suburb'] == suburb]
        bar_data = filtered_df['ACCIDENT_TYPE_DESC'].value_counts().reset_index()
        bar_data.columns = ['ACCIDENT_TYPE_DESC', 'COUNT']
        
        bar_fig = px.bar(
            bar_data,
            x='ACCIDENT_TYPE_DESC',
            y='COUNT',
            color='COUNT',
            color_continuous_scale='Blues',
            title=f"Accident Types in {suburb}",
            labels={'ACCIDENT_TYPE_DESC': 'Accident Type', 'COUNT': 'Number of Accidents'}
        )
        
        bar_fig.update_layout(
            title={'x':0.5, 'xanchor': 'center'},
            xaxis_title="Accident Type",
            yaxis_title="Number of Accidents",
            template='plotly_white'
        )
        return bar_fig
    
    return go.Figure()

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```
